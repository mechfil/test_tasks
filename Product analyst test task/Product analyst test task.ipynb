```python
# приводим типы данных

users['reg_time'] = pd.to_datetime(users['reg_time'], 
                                   format='%Y-%m-%d %H:%M:%S.%f')
sessions['close_time'] = pd.to_datetime(sessions['close_time'], 
                                   format='%Y-%m-%d %H:%M:%S')
sessions['open_time'] = pd.to_datetime(sessions['open_time'], 
                                   format='%Y-%m-%d %H:%M:%S')
payments['pay_time'] = pd.to_datetime(payments['pay_time'], 
                                   format='%Y-%m-%d %H:%M:%S.%f')
levels['time'] = pd.to_datetime(levels['time'], 
                                   format='%Y-%m-%d %H:%M:%S.%f')
levels['level'] = levels['level'].astype('int64')
```

### Проверка времени регистраций

Проверим, есть ли пользователи, у которых момент первого посещения происходил до момента регистрации. Такого не должно быть.

```python
# находим моменты первых посещений
sessions_first = sessions.groupby('user_id').agg({'open_time': 'first'})

# склеиваем таблицу регистраций и первых посещений
users_sessions_first = users.merge(sessions_first, on='user_id', how='left')

# отбираем строки с моментами регистрации после первых посещений
users_defective = users_sessions_first[users_sessions_first['reg_time'] > users_sessions_first['open_time']]

# рассчитываем разницу между моментом регистрации и первого посещения
users_defective['diff'] = (
    users_defective['open_time'] - users_defective['reg_time']
).dt.total_seconds() / 60

# выводим таблицу дефективных пользователей
users_defective.sort_values(by='diff')
```

**Вывод:**
- Есть пользователи, у которых момент регистрации происходил после первого посещения. У почти всех таких пользователей разница между этими событиями не превышает 20 минут. Считаю, что это можно списать на погрешности записи и заменить дату регистрации у таких пользователей датой первого посещения.

Заменим в таблице регистраций `users` дату и время регистрации на дату и время первого посещения у тех пользователей, у которых момент регистрации происходил после первого посещения.

```python
# заменяем момент регистрации моментом первого посещения, если регистрации произошла после первого посещения
users_sessions_first.loc[users_sessions_first['open_time'] 
                         < users_sessions_first['reg_time'], 'reg_time'] = users_sessions_first['open_time']

# заменяем таблицу регистраций новыми данными
users = users_sessions_first[['user_id','reg_time']]
```

Зададим собственную функцию для визуализации распределений.

```python
# функция для построения точечного графика, боксплота и вывода основых статистик

def describe_func(df,                               # датафрейм
                  column,                           # колонка с данными
                  title = 'Распределение значений', # заголовок графика по умолчанию
                  group_column='all',               # колонка для группировки
                 ):
    
    # копируем датафрейм
    df = df.copy()
    
    # если нет колонки для группировки создаем колонку 'all'
    if group_column == 'all':
        df['all'] = 'all'
    
    # задаем счетчик
    i = 0
    
    # создаем таблицу со статистиками
    for group_name in sorted(df[group_column].unique()):
        
        # получаем основные статистики
        df_desc = pd.DataFrame(df[df[group_column] == group_name][column].describe())
        
        # если есть группировка, переименовываем колонку
        if group_column != 'all':
            df_desc = df_desc.rename(columns={column: column+'_'+str(group_name)})
        
        # считаем значения усов боксплота
        quant_1 = df[df[group_column] == group_name][column].quantile([0.25])[0.25]
        quant_3 = df[df[group_column] == group_name][column].quantile([0.75])[0.75]
        IQR = quant_3 - quant_1
        y_min = quant_1 - 1.5 * IQR
        y_max = quant_3 + 1.5 * IQR
        
        # записываем диапазоны осей для графика
        if i == 0:
            y_min_box = y_min
            y_max_box = y_max
        else:
            if y_min < y_min_box:
                y_min_box = y_min
            if y_max > y_max_box:
                y_max_box = y_max
        
        # склеиваем данные в одну таблицу
        if i == 0:
            df_desc_all = df_desc
        else:
            df_desc_all = df_desc_all.join(df_desc)
            
        # увеличиваем счетчик
        i += 1
    
    # выводим таблицу со статистиками
    display(df_desc_all.round(2))
    
    # задаем единый стиль графиков
    sns.set_style('darkgrid')
    
    # задаем размер подложки для графиков
    fig = plt.subplots(figsize=(14,6))
    
    # строим график-полоску
    ax1 = plt.subplot(1, 2, 1)
    ax1 = sns.stripplot(y=column, x=group_column, data=df, color='salmon')
    ax1.set_title(title + '\n')
    
    # строим график-полоску и боксплот в увеличенном масштабе
    ax2 = plt.subplot(1, 2, 2)
    ax2 = sns.stripplot(y=column, x=group_column, data=df, color='salmon')
    ax2 = sns.boxplot(y=column, x=group_column, data=df, color='limegreen', showfliers = False)
    ax2.set_title(title + '\n' + '(увеличенный масштаб)\n')
       
    # немного сдвигаем диапазоны осей при отображении
    if df[df[column] < y_min_box][column].count() != 0:
        y_min_ax = y_min_box-0.2*abs(y_min_box)
    else:
        y_min_ax = df[column].min()-(y_max_box-y_min_box)*0.05
    if df[df[column] > y_max_box][column].count() != 0:
        y_max_ax = y_max_box+0.2*abs(y_max_box)
    else:
        y_max_ax = df[column].max()+(y_max_box-y_min_box)*0.05
    
    # ограничиваем оси
    ax2.set_ylim(y_min_ax, y_max_ax)

    # если нет групировки, убираем надписи
    if group_column == 'all':
        ax1.set_xlabel('')
        ax1.set_xticklabels([])
        ax2.set_xlabel('')
        ax2.set_xticklabels([])
    
    # выводим графики
    plt.show()
```

### Расчет и визуализация удержания

Вначале подготовим таблицу профилей учеников.

```python
# функция для создания пользовательских профилей

def get_profiles(
    sessions,       # таблица посещений (или регистраций)
    payments,       # таблица покупок
    user_id,        # столбец с id пользователей в таблице посещений (регистраций) и таблице покупок
    session_time,   # столбец со временем посещений
):

    # находим параметры первых посещений (в таблице регистраций ничего не изменится)
    profiles = (
        sessions.sort_values(by=[user_id, session_time])
        .groupby(user_id)
        .agg(
            {
                session_time: 'first'
                # сюда также можно добавить столбцы с доп. признаками вида:
                # 'имя столбца': 'first'
            }
        )
        .rename(columns={session_time: 'first_ts'})
        .reset_index()
    )
    
    # добавляем признак платящих пользователей
    profiles['payer'] = profiles['user_id'].isin(payments['user_id'].unique())

    return profiles
```

```python
# создаем профили пользователей
profiles = get_profiles(
    users,
    payments,
    'user_id',
    'reg_time',
)
```

Подготовим вспомогательную функцию для расчета удержания

```python
# вспомогательная функция для расчёта удержания

def get_retention(
    profiles,           # таблица профилей
    sessions,           # таблица посещений
    user_id,            # столбец с id клиентов в таблице посещений и профилей
    reg_time,           # столбец с датой регистраций (или первых посещений) в таблице профилей
    session_time,       # столбец с датой посещений в таблице посещений
    observation_date,   # момент анализа
    horizon_days,       # горизонт анализа
    dimensions,         # группировка по желаемым признакам
    ignore_horizon,     # игнорировать ли горизонт
):

    # проверяем момент анализа
    if observation_date > sessions[session_time].max():
        print('Внимание! Заданный вами момент анализа превышает максимальную дату посещений. '
              'Значение момента анализа уменьшено до максимально возможного: {}.\n'
              .format(sessions[session_time].max().date()))
        observation_date = sessions[session_time].max().date()
    
    # добавляем дату первых посещений в таблицу профилей
    profiles.loc[:,'dt'] = profiles[reg_time].dt.date
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
        # если нет пользователей, подпадающих под параметры, выводим сообщение об ошибке
        if profiles['dt'].min() > last_suitable_acquisition_date:
            raise Exception('Ошибка! Нет данных, подпадающих под заданные в функции параметры. '
                            'Рекомендуется уменьшить значение горизонта анализа '
                            'и/или увеличить значение момента анализа.')
    # оставляем только нужных пользователей
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # добавляем данные о посещениях в профили
    result_raw = result_raw.merge(
        sessions[[user_id, session_time]], on=user_id, how='left'
    )
    
    # рассчитываем лайфтайм пользователя для каждой посещения
    result_raw['lifetime'] = (
        result_raw[session_time] - result_raw[reg_time]
    ).dt.days.astype('Int64')
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days, k=[0]):
        # строим «треугольную» таблицу посещений
        result = df.pivot_table(
            index=dims, columns='lifetime', values=user_id, aggfunc='nunique'
        )
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({user_id: 'nunique'})
            .rename(columns={user_id: 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу посещений
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем удержание: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # удаляем столбец с размерами когорт
        result = result.drop(columns=['cohort_size'])
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result.loc[:, (result.columns < horizon_days)]
        # находим лайфтаймы, за которые нет данных
        set_1 = set(result.columns.astype(int))
        list_all = sorted(list(set_1))
        set_1 = set(filter(lambda x: x >= 0, set_1))
        set_2 = set(range(result.columns.astype(int).max()+1))
        list_nan = sorted(list(set_1 ^ set_2))
        if len(list_nan) > 0:
            # заполняем пропуски в лайфтамах значениями из предыдущих столбцов
            def insert_number(lst, num, df_in):
                index = len(lst)
                for i, n in enumerate(lst):
                    if n > num:
                        index = i
                        break
                    value = n    
                df_in.insert(index, num, df_in[value])
                return df_in
            for lt in list_nan:
                result = insert_number(result.columns.astype(int), lt, result)
            # выводим предупреждение (только если функция вызывается впервые)
            k[0] += 1
            if k[0] == 1:
                list_format = ', '.join(str(x) for x in list_nan)
                print('Внимание! За эти лайфтаймы нет данных: {}'.format(list_format))
                print('Пропуски заполнены значениями из предыдущих лайфтаймов.\n')
        # восстанавливаем столбец с размерами когорт
        result.insert(0, 'cohort_size', cohort_sizes)
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )
    
    # для таблицы динамики удержания убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )
    
    # получаем таблицу динамики LTV для хитмепа (без группировок по признакам)
    result_heatmap = group_by_dimensions(
        result_raw, ['dt'], horizon_days
    )
        
    # возвращаем таблицы и сырые данные
    return (
        result_raw,       # сырые данные
        result_grouped,   # таблица удержания
        result_in_time,   # таблица динамики удержания
        result_heatmap    # таблица динамики удержания для хитмпепа
    )   
```

Подготовим вспомогательную функцию для визуализации удержания

```python
# вспомогательная функция для визуализации удержания

def plot_retention(
    retention,             # таблица удержания
    retention_history,     # таблица динамики удержания
    retention_heatmap,     # таблица динамики удержания для хитмепа
    horizon,               # горизонт анализа
    dimensions,            # группировка по желаемым признакам
    ex_lifetimes,          # какие лайфтаймы следует исключить
    heatmap_on,            # нужно ли также строить хитмеп удержания по когортам дней регистрации
    comm_axis,             # сделать ли общую ось у двух графиков
    title,                 # заголовок графиков
    window,                # параметр сглаживания графика динамики удержания
    heatmap_size,          # размер тепловой карты
    heatmap_annot          # включить значения на тепловой карте
):

    # функция для сглаживания фрейма
    def filter_data(df, window):
        for column in df.columns.values:
            df[column] = df[column].rolling(window).mean() 
        return df
    
    # задаём размер сетки для графиков
    plt.figure(figsize=(12, 5))
    
    # задаем стиль графиков
    sns.set_style('darkgrid')

    # из таблицы удержания исключаем размеры когорт
    retention = retention.drop(columns=['cohort_size'])
    # если требуется, исключаем заданные лайфтаймы
    if len(ex_lifetimes) != 0:
        retention = retention.drop(columns=ex_lifetimes)
        retention_history = retention_history.drop(columns=ex_lifetimes)
    # в таблице динамики удержания оставляем только нужный лайфтайм
    retention_history = retention_history.drop(columns=['cohort_size'])
    if retention_history.columns.astype(int).max() == (horizon - 1):
        retention_history = retention_history[[horizon - 1]]
    else:
        # если нужного лайфтайма нет, выводим предупреждение
        print('Внимание! Нет данных за {} лайфтайм. ' 
              'Графики построены до максимального {} лайфтайма.\n'
              .format((horizon - 1),retention_history.columns.astype(int).max()))
    
    # строим график удержания
    ax1 = plt.subplot(1, 2, 1)
    retention.T.plot(
        grid=True, ax=ax1, marker='o'
    )
    plt.xlabel('Лайфтайм')
    plt.title(title)
    plt.gca().yaxis.set_major_formatter('{x:.2%}')
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()

    # строим график динамики удержания
    if comm_axis == False:
        ax2 = plt.subplot(1, 2, 2)
    else:
        ax2 = plt.subplot(1, 2, 2, sharey = ax1)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['dt']]
    # фильтруем данные и строим график
    filtered_data = retention_history.pivot_table(
        index='dt', columns=columns, values=retention_history.columns.astype(int).max(), aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2, marker='o', rot=45)
    plt.gca().yaxis.set_major_formatter('{x:.2%}')
    plt.xlabel('Дата регистрации')
    plt.title('{}: динамика на {}-й день'.format(title,retention_history.columns.astype(int).max()+1))
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()

    # выводим оба графика
    plt.tight_layout()
    plt.show()
    
    # если нужно, также строим хитмеп удержания по когортам дней регистрации
    if heatmap_on != False:
        plt.figure(figsize = (12, retention_heatmap.shape[0]/heatmap_size))
        # если требуется, исключаем заданные лайфтаймы
        if len(ex_lifetimes) != 0:
            retention_heatmap = retention_heatmap.drop(columns=ex_lifetimes)
        sns.heatmap(retention_heatmap.drop(columns=['cohort_size']),
                annot = heatmap_annot,
                fmt = '.2%', 
                cmap= 'coolwarm',
                linecolor='black',
                linewidths=0.01
               )
        plt.title('{}: тепловая карта'.format(title))
        plt.ylabel('Дата регистрации')
        plt.xlabel('Лайфтайм')
        plt.show()
```

Подготовим главную функцию для расчета и визуализации удержания

```python
# главная функция расчета и визуализации удержания

'''
Перед использованием функции небходимо подготовить две таблицы:
1. Таблица профилей (profiles) вида: 
id клиента (user_id) - дата и время регистрации / первого посещения (reg_time) - доп. атрибуты (dimensions)
2. Таблица посещений (sessions) вида:
id клиента (user_id) - дата и время посещения (session_time)
'''

def retention_final(
    profiles,               # таблица профилей
    sessions,               # таблица посещений
    user_id,                # столбец с id клиентов в таблице посещений и профилей
    reg_time,               # столбец с датой регистраций (или первых посещений) в таблице профилей
    session_time,           # столбец с датой посещений в таблице посещений
    observation_date,       # момент анализа
    horizon_days,           # горизонт анализа 
    dimensions = [],        # группировка по желаемым признакам
    ex_lifetimes = [],      # исключение желаемых лайфтаймов
    comm_axis = False,      # сделать ли общую ось у двух графиков
    title = 'Удержание',    # имя переменной в заголовках графиков
    window = 1,             # параметр сглаживания графика динамики удержания
    ignore_horizon = False, # игнорировать ли горизонт
    heatmap_on = False,     # нужно ли также построить хитмеп удержания по когортам дней регистрации
    heatmap_size = 3,       # задаем размер тепловой карты
    heatmap_annot = True    # включить ли подписи на тепловой карте
):
    
    # получаем таблицы удержания
    (
        retention_raw,
        retention_grouped,
        retention_history,
        retention_heatmap
    ) = get_retention(
        profiles,
        sessions,
        user_id,
        reg_time,
        session_time,
        observation_date,
        horizon_days,
        dimensions,
        ignore_horizon
        
    )

    # строим графики удержания
    plot_retention(
        retention_grouped,
        retention_history,
        retention_heatmap,
        horizon_days,
        dimensions,
        ex_lifetimes,
        heatmap_on,
        comm_axis,
        title,
        window,
        heatmap_size,
        heatmap_annot
    )
```

### Расчет и визуализации времени в приложении

Для подсчета времени в приложении используем собственную функцию LTV (мы будем использовать ее ниже для расчета LTV), но вместо выручки здесь зададим время, проведенное в этот день в приложении.

Подготовим вспомогательную функцию для расчета LTV

```python
# вспомогательная функция для расчёта LTV

def get_ltv(
    profiles,         # таблица профилей
    payments,         # таблица покупок
    user_id,          # столбец с id клиентов в таблице профилей и покупок
    reg_time,         # столбец с датами регистраций (или первых посещений) в таблице профилей
    pay_time,         # столбец с датами покупок в таблице покупок
    pay_value,        # столбец с суммой покупок в таблице покупок
    observation_date, # момент анализа
    horizon_days,     # горизонт анализа
    dimensions,       # группировка по желаемым признакам
    ignore_horizon    # игнорировать ли горизонт
):
    # проверяем момент анализа
    if observation_date > payments[pay_time].max():
        print('Внимание! Заданный вами момент анализа превышает максимальную дату покупок. '
              'Значение момента анализа уменьшено до максимально возможного: {}.\n'
              .format(payments[pay_time].max().date()))
        observation_date = payments[pay_time].max().date()
    
    # добавляем дату первых посещений в таблицу профилей
    profiles['dt'] = profiles[reg_time].dt.date
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
        # если нет пользователей, подпадающих под параметры, выводим сообщение об ошибке
        if profiles['dt'].min() > last_suitable_acquisition_date:
            raise Exception('Ошибка! Нет данных, подпадающих под заданные в функции параметры. '
                            'Рекомендуется уменьшить значение горизонта анализа '
                            'и/или увеличить значение момента анализа.')
    # оставляем только нужных пользователей
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    
    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        payments[[user_id, pay_time, pay_value]], on=user_id, how='left'
    )
    
    # рассчитываем лайфтайм пользователя для каждой покупки
    result_raw['lifetime'] = (
        result_raw[pay_time] - result_raw[reg_time]
    ).dt.days.astype('Int64')
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция группировки по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days, k=[0]):
        # строим «треугольную» таблицу выручки
        result = df.pivot_table(
            index=dims, columns='lifetime', values=pay_value, aggfunc='sum'
        )
        # находим сумму выручки с накоплением
        result = result.fillna(0).cumsum(axis=1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({user_id: 'nunique'})
            .rename(columns={user_id: 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу выручки
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем LTV: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # удаляем столбец с размерами когорт
        result = result.drop(columns=['cohort_size'])
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result.loc[:, (result.columns < horizon_days)]
        # находим лайфтаймы, за которые нет данных
        set_1 = set(result.columns.astype(int))
        list_all = sorted(list(set_1))
        set_1 = set(filter(lambda x: x >= 0, set_1))
        set_2 = set(range(result.columns.astype(int).max()+1))
        list_nan = sorted(list(set_1 ^ set_2))
        if len(list_nan) > 0:
            # заполняем пропуски в лайфтамах значениями из предыдущих столбцов
            def insert_number(lst, num, df_in):
                index = len(lst)
                for i, n in enumerate(lst):
                    if n > num:
                        index = i
                        break
                    value = n    
                df_in.insert(index, num, df_in[value])
                return df_in
            for lt in list_nan:
                result = insert_number(result.columns.astype(int), lt, result)
            # выводим предупреждение (только если функция вызывается впервые)
            k[0] += 1
            if k[0] == 1:
                list_format = ', '.join(str(x) for x in list_nan)
                print('Внимание! За эти лайфтаймы нет данных: {}'.format(list_format))
                print('Пропуски заполнены значениями из предыдущих лайфтаймов.\n')
        # восстанавливаем столбец с размерами когорт
        result.insert(0, 'cohort_size', cohort_sizes)
        return result
    
    # получаем таблицу LTV
    result_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблицы динамики LTV убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицу динамики LTV
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )
    
    # получаем таблицу динамики LTV для хитмепа (без группировок по признакам)
    result_heatmap = group_by_dimensions(
        result_raw, ['dt'], horizon_days
    )

    return (
        result_raw,       # сырые данные
        result_grouped,   # таблица LTV
        result_in_time,   # таблица динамики LTV
        result_heatmap    # таблица динамики LTV для хитмпепа
    )
```

Подготовим вспомогательную функцию для визуализации LTV

```python
# вспомогательная функция для визуализации LTV

def plot_ltv(
    ltv,           # таблица LTV
    ltv_history,   # таблица динамики LTV
    ltv_heatmap,   # таблица динамики LTV для хитмепа
    horizon,       # горизонт анализа
    dimensions,    # группировка по желаемым признакам
    ex_lifetimes,  # какие лайфтаймы следует исключить
    heatmap_on,    # нужно ли также строить хитмеп LTV по когортам дней регистрации
    comm_axis,     # сделать ли общую ось у двух графиков
    title,         # заголовок графиков
    window,        # параметр сглаживания графика динамики удержания
    heatmap_size,  # размер тепловой карты
    heatmap_annot  # включить ли значения на тепловой карте
):

    # функция для сглаживания фрейма
    def filter_data(df, window):
        for column in df.columns.values:
            df[column] = df[column].rolling(window).mean() 
        return df
    
    # задаём сетку отрисовки графиков
    plt.figure(figsize=(12, 5))
    
    # задаем стиль графиков
    sns.set_style('darkgrid')

    # из таблицы ltv исключаем размеры когорт
    ltv = ltv.drop(columns=['cohort_size'])
    # если требуется, исключаем заданные лайфтаймы
    if len(ex_lifetimes) != 0:
        ltv = ltv.drop(columns=ex_lifetimes)
        ltv_history = ltv_history.drop(columns=ex_lifetimes)
    # в таблице динамики ltv оставляем только нужный лайфтайм
    ltv_history = ltv_history.drop(columns=['cohort_size'])
    if ltv_history.columns.astype(int).max() == (horizon - 1):
        ltv_history = ltv_history[[horizon - 1]]
    else:
        # если нужного лайфтайма нет, выводим предупреждение
        print('Внимание! Нет данных за {} лайфтайм. ' 
              'Графики построены до максимального {} лайфтайма.\n'
              .format((horizon - 1),ltv_history.columns.astype(int).max()))
        
    # первый график — кривые ltv
    ax1 = plt.subplot(1, 2, 1)
    ltv.T.plot(grid=True, ax=ax1, marker='o')
    plt.xlabel('Лайфтайм')
    plt.title(title)
    if len(dimensions) == 0:
        plt.legend().set_visible(False)
    else:
        plt.legend()

    # второй график — динамика ltv
    if comm_axis == False:
        ax2 = plt.subplot(1, 2, 2)
    else:
        ax2 = plt.subplot(1, 2, 2, sharey = ax1)
    # столбцами сводной таблицы станут все столбцы индекса, кроме даты
    columns = [name for name in ltv_history.index.names if name not in ['dt']]
    filtered_data = ltv_history.pivot_table(
        index='dt', columns=columns, values=ltv_history.columns.astype(int).max(), aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2, marker='o', rot=45)
    plt.xlabel('Дата регистрации')
    plt.title('{}: динамика на {}-й день'.format(title, ltv_history.columns.astype(int).max()+1))
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()
        
    # выводим оба графика    
    plt.tight_layout()
    plt.show()
    
    # если нужно, также строим хитмеп LTV по когортам дней регистрации
    if heatmap_on != False:
        plt.figure(figsize = (12, ltv_heatmap.shape[0]/heatmap_size))
        # если требуется, исключаем заданные лайфтаймы
        if len(ex_lifetimes) != 0:
            ltv_heatmap = ltv_heatmap.drop(columns=ex_lifetimes)
        sns.heatmap(ltv_heatmap.drop(columns=['cohort_size']),
                annot = heatmap_annot,
                fmt = '.2f', 
                cmap= 'coolwarm',
                linecolor='black',
                linewidths=0.01
               )
        plt.title('{}: тепловая карта'.format(title))
        plt.ylabel('Дата регистрации')
        plt.xlabel('Лайфтайм')
        plt.show()
```

Зададим главную функцию для расчета и визуализации LTV

```python
# главная функция расчета и визуализации LTV

'''
Перед использованием функции небходимо подготовить две таблицы:
1. Таблица профилей (profiles) вида: 
id клиента (user_id) - дата и время регистрации / первого посещения (reg_time) - доп. атрибуты (dimensions)
2. Таблица покупок (payments) вида:
id клиента (user_id) - дата и время покупки (pay_time) - сумма покупки (pay_value)
'''

def ltv_final(
    profiles,               # таблица профилей
    payments,               # таблица покупок
    user_id,                # столбец с id клиентов в таблице покупок и профилей
    reg_time,               # столбец с датой регистраций (или первых посещений) в таблице профилей
    pay_time,               # столбец с датой покупок в таблице покупок
    pay_value,              # столбец с суммой покупок в таблице покупок
    observation_date,       # момент анализа
    horizon_days,           # горизонт анализа 
    dimensions = [],        # группировка по желаемым признакам
    ex_lifetimes = [],      # исключение желаемых лайфтаймов
    comm_axis = False,      # сделать ли общую ось у двух графиков
    title = 'LTV',          # имя переменной в заголовках графиков
    window = 1,             # параметр сглаживания графика динамики LTV
    ignore_horizon = False, # игнорировать ли горизонт
    heatmap_on = False,     # нужно ли также построить хитмеп LTV по когортам дней регистрации
    heatmap_size = 3,       # задаем размер тепловой карты
    heatmap_annot = True    # включить ли подписи на тепловой карте
):
        
    # получаем таблицы LTV
    (
        ltv_raw,
        ltv_grouped, 
        ltv_history,
        ltv_heatmap
    ) = get_ltv(
        profiles,
        payments,
        user_id,
        reg_time,
        pay_time,
        pay_value,
        observation_date,
        horizon_days,
        dimensions,
        ignore_horizon
    )

    # строим графики LTV
    plot_ltv(
        ltv_grouped, 
        ltv_history,
        ltv_heatmap,
        horizon_days,
        dimensions,
        ex_lifetimes,
        heatmap_on,
        comm_axis,
        title,
        window,
        heatmap_size,
        heatmap_annot
    )
```

```python
# считаем время сессий
sessions_time = sessions.copy()
sessions_time['session_date'] = sessions_time['open_time']
sessions_time['session_time'] = (sessions_time['close_time'] -  sessions_time['open_time'])

# оставляем нужные столбцы
sessions_time = sessions_time[['user_id','session_date','session_time']]

# выведем отсортированный датафрейм по времени сессии
sessions_time.sort_values(by='session_time')
```

Стоит отметить, что в нашем решении **время в приложении привязано к началу сессии**. То есть если ученик зарегистрировался 23 часа назад, сейчас зашел в приложение и провел в нем 3 часа, то все 3 часа пойдут в нулевой лайфтайм (один лайфтайм = 24 часа), а не 1 час - в нулевой лайфтайм и еще 2 часа - в следующий первый лайфтайм. При этом, если этот ученик сейчас зашел в приложение, провел в нем 1 час, затем вышел из него, зашел снова и провел в нем еще 2 часа, то 1 час времени в приложении пойдет в нулевой лайфтайм, а 2 часа - в первый, так как эти 2 часа будут уже привязаны к началу следующей сессии, которая началась спустя 24 часа.

Переведем столбец в минуты.

```python
sessions_time['session_time'] = sessions_time['session_time'].dt.total_seconds() / 60
```

Посчитаем время, проведенное на уровнях. Посчитаем как время нахождения (то есть время, которое в среднем провели на уровне ученики, дошедшие до уровня), так и время прохождения (то есть время, которое в среднем потратили ученики, полностью прошедшие уровень).

```python
# приведем таблицу посещений к вертикальной, добавим флаги событий
s = (sessions.melt(
    id_vars=['user_id'], value_vars=['close_time', 'open_time'])
      .rename(columns={'variable': 'flag','value':'time'})[['user_id','time','flag']])

# переименуем колонку в таблице с уровнями
l = levels.rename(columns={'level_time': 'time'})

# добавим флаг уровня в таблицу уровней
l['flag'] = 'level_time'

# объединяем вертикально таблицы посещений и уровней, сортируем по id учеников и времени события
s_l = pd.concat([s, l], ignore_index=True).sort_values(by=['user_id','time'])

# добавляем время, проведенное между событиями, для каждого пользователя в минутах
s_l['time_delta'] = s_l.groupby('user_id')['time'].diff().shift(-1).dt.total_seconds() / 60

# удаляем все строки с событием close_time
s_l = s_l[s_l['flag'] != 'close_time']

# заполняем промежутки между уровнями
s_l['level'] = s_l.groupby('user_id')['level'].fillna(method='ffill')

# заполняем пропуски в уровнях 1-м уровнем для событий, которые происходили до 2 уровня у каждого ученика
s_l['level'] = s_l['level'].fillna(1.0).astype(int)

# находим суммарное количество времени от всех учеников на уровнях
lev_time = s_l.groupby('level')['time_delta'].sum()

# находим суммарное количество учеников на уровнях
users_count = s_l.groupby('level')['user_id'].nunique()

# находим среднее время, проведенного одним учеником, на каждом уровне
mean_time_lev = lev_time / users_count

# посчитаем теперь только время на уровнях среди пользователей,
# которые полностью прошли конкретный уровень

# группируем таблицу по пользователям и уровням
s_l_buf = s_l.groupby(['user_id','level'])['time_delta'].sum().reset_index()

# удаляем для каждого пользователя последний уровень, на котором он остановился
s_l_buf = s_l_buf.groupby('user_id').apply(lambda x: x.drop(x.last_valid_index())).reset_index(drop=True)

# считаем среднее время прохождения каждого уровня
mean_time_lev_finish = s_l_buf.groupby('level')['time_delta'].mean()

# объединяем данные для графика
time_lev = (pd.DataFrame(mean_time_lev).rename(columns={0: 'mean_time_lev'})
       .join(pd.DataFrame(mean_time_lev_finish).rename(columns={'time_delta': 'mean_time_lev_finish'})))
time_lev = time_lev.reset_index()
time_lev = (time_lev.melt(
    id_vars=['level'], value_vars=['mean_time_lev', 'mean_time_lev_finish']).rename(columns={'value': 'time'}))

# строим график
plt.figure(figsize=(10, 5))
sns.barplot(x='level', y='time', hue='variable', data=time_lev)
plt.title('Среднее время нахождения на уровнях (mean_time_lev) \n и среднее время прохождения уровней (mean_time_lev_finish) в минутах')
plt.show();
```

**Выводы:**
- До 7 уровня ученики относительно быстро проходят уровни: время прохождения одного уровня занимает меньше 50 минут.
- Начиная с 8 уровня время прохождения начинает резко возрастать: время прохождения одного уровня становится выше 100 минут.

#### Расчет и визуализация конверсии

Подготовим вспомогательную функцию для расчета конверсии

```python
# вспомогательная функция для расчёта конверсии

def get_conversion(
    profiles,         # таблица профилей
    payments,         # таблица покупок
    user_id,          # столбец с id клиентов в таблице покупок и профилей
    reg_time,         # столбец с датами регистраций (или первых посещений) в таблице профилей
    pay_time,         # столбец с датами покупок в таблице покупок
    observation_date, # момент анализа
    horizon_days,     # горизонт анализа
    dimensions,       # группировка по желаемым признакам
    ignore_horizon,   # игнорировать ли горизонт
):

    # проверяем момент анализа
    if observation_date > payments[pay_time].max():
        print('Внимание! Заданный вами момент анализа превышает максимальную дату покупок. '
              'Значение момента анализа уменьшено до максимально возможного: {}.\n'
              .format(payments[pay_time].max().date()))
        observation_date = payments[pay_time].max().date()

    # добавляем дату первых посещений в таблицу профилей
    profiles['dt'] = profiles[reg_time].dt.date
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
        # если нет пользователей, подпадающих под параметры, выводим сообщение об ошибке
        if profiles['dt'].min() > last_suitable_acquisition_date:
            raise Exception('Ошибка! Нет данных, подпадающих под заданные в функции параметры. '
                            'Рекомендуется уменьшить значение горизонта анализа '
                            'и/или увеличить значение момента анализа.')
    # оставляем только нужных пользователей
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    
    # определяем дату и время первой покупки для каждого пользователя
    first_purchases = (
        payments.sort_values(by=[user_id, pay_time])
        .groupby(user_id)
        .agg({pay_time: 'first'})
        .reset_index()
    )

    # добавляем данные о покупках в профили
    result_raw = result_raw.merge(
        first_purchases[[user_id, pay_time]], on=user_id, how='left'
    )

    # рассчитываем лайфтайм для каждой покупки
    result_raw['lifetime'] = (
        result_raw[pay_time] - result_raw[reg_time]
    ).dt.days.astype('Int64')

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days, k=[0]):
        # строим «треугольную» таблицу конверсии
        result = df.pivot_table(
            index=dims, columns='lifetime', values=user_id, aggfunc='nunique'
        )
        # находим конверсию с накоплением
        result = result.fillna(0).cumsum(axis = 1)
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({user_id: 'nunique'})
            .rename(columns={user_id: 'cohort_size'})
        )
        # объединяем размеры когорт и таблицу конверсии
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # считаем конверсию: делим каждую «ячейку» в строке на размер когорты
        result = result.div(result['cohort_size'], axis=0)
        # удаляем столбец с размерами когорт
        result = result.drop(columns=['cohort_size'])
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result.loc[:, (result.columns < horizon_days)]
        # находим лайфтаймы, за которые нет данных
        set_1 = set(result.columns.astype(int))
        list_all = sorted(list(set_1))
        set_1 = set(filter(lambda x: x >= 0, set_1))
        set_2 = set(range(result.columns.astype(int).max()+1))
        list_nan = sorted(list(set_1 ^ set_2))
        if len(list_nan) > 0:
            # заполняем пропуски в лайфтамах значениями из предыдущих столбцов
            def insert_number(lst, num, df_in):
                index = len(lst)
                for i, n in enumerate(lst):
                    if n > num:
                        index = i
                        break
                    value = n    
                df_in.insert(index, num, df_in[value])
                return df_in
            for lt in list_nan:
                result = insert_number(result.columns.astype(int), lt, result)
            # выводим предупреждение (только если функция вызывается впервые)
            k[0] += 1
            if k[0] == 1:
                list_format = ', '.join(str(x) for x in list_nan)
                print('Внимание! За эти лайфтаймы нет данных: {}'.format(list_format))
                print('Пропуски заполнены значениями из предыдущих лайфтаймов.\n') 
        # восстанавливаем столбец с размерами когорт
        result.insert(0, 'cohort_size', cohort_sizes)        
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )
    
    # если нужно, также строим таблицу конверсии для хитмепа (без группировок по признакам)
    result_heatmap = group_by_dimensions(
        result_raw, ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return (
        result_raw,       # сырые данные
        result_grouped,   # таблица конверсии
        result_in_time,   # таблица динамики конверсии
        result_heatmap  # таблица динамики конверсии для хитмпепа
    )
```

Подготовим вспомогательную функцию для визуализации конверсии

```python
# вспомогательная функция для визуализации конверсии

def plot_conversion(
    conversion,            # таблица конверсии
    conversion_history,    # таблица динамики конверсии
    conversion_heatmap,    # таблица динамики конверсии для хитмепа
    horizon,               # горизонт анализа
    dimensions,            # группировка по желаемым признакам
    ex_lifetimes,          # какие лайфтаймы следует исключить
    heatmap_on,            # нужно ли также строить хитмеп конверсий по когортам дней регистрации
    comm_axis,             # сделать ли общую ось у двух графиков
    title,                 # заголовок графиков
    window,                # параметр сглаживания графика динамики конверсии
    heatmap_size,          # размер тепловой карты
    heatmap_annot          # включить значения на тепловой карте
):

    # функция для сглаживания фрейма
    def filter_data(df, window):
        for column in df.columns.values:
            df[column] = df[column].rolling(window).mean() 
        return df
    
    # задаём размер сетки для графиков
    plt.figure(figsize=(12, 5))
    
    # задаем стиль графиков
    sns.set_style('darkgrid')

    # из таблицы конверсии исключаем размеры когорт
    conversion = conversion.drop(columns=['cohort_size'])
    # если требуется, исключаем заданные лайфтаймы
    if len(ex_lifetimes) != 0:
        conversion = conversion.drop(columns=ex_lifetimes)
        conversion_history = conversion_history.drop(columns=ex_lifetimes)
    # в таблице динамики конверсии оставляем только нужный лайфтайм
    conversion_history = conversion_history.drop(columns=['cohort_size'])
    if conversion_history.columns.astype(int).max() == (horizon - 1):
        conversion_history = conversion_history[[horizon - 1]]
    else:
        # если нужного лайфтайма нет, выводим предупреждение
        print('Внимание! Нет данных за {} лайфтайм. ' 
              'Графики построены до максимального {} лайфтайма.\n'
              .format((horizon - 1),conversion_history.columns.astype(int).max()))
        
    # первый график — кривые конверсии
    ax1 = plt.subplot(1, 2, 1)
    conversion.T.plot(grid=True, ax=ax1, marker='o')
    plt.gca().yaxis.set_major_formatter('{x:.2%}')
    plt.xlabel('Лайфтайм')
    plt.title(title)
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()

    # второй график — динамика конверсии
    if comm_axis == False:
        ax2 = plt.subplot(1, 2, 2)
    else:
        ax2 = plt.subplot(1, 2, 2, sharey = ax1)
    columns = [
        # столбцами сводной таблицы станут все столбцы индекса, кроме даты
        name for name in conversion_history.index.names if name not in ['dt']
    ]
    filtered_data = conversion_history.pivot_table(
        index='dt', columns=columns, values=conversion_history.columns.astype(int).max(), aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2, marker='o', rot=45)
    plt.xlabel('Дата регистрации')
    plt.title('{}: динамика на {}-й день'.format(title,conversion_history.columns.astype(int).max()+1))
    plt.gca().yaxis.set_major_formatter('{x:.2%}')
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()

    # выводим оба графика 
    plt.tight_layout()
    plt.show()
    
    # если нужно, также строим хитмеп конверсии по когортам дней регистрации
    if heatmap_on != False:
        plt.figure(figsize = (12, conversion_heatmap.shape[0]/heatmap_size))
        # если требуется, исключаем заданные лайфтаймы
        if len(ex_lifetimes) != 0:
            conversion_heatmap = conversion_heatmap.drop(columns=ex_lifetimes)
        sns.heatmap(conversion_heatmap.drop(columns=['cohort_size']),
                annot = heatmap_annot,
                fmt = '.2%', 
                cmap= 'coolwarm',
                linecolor='black',
                linewidths=0.01
               )
        plt.title('{}: тепловая карта'.format(title))
        plt.ylabel('Дата регистрации')
        plt.xlabel('Лайфтайм')
        plt.show()
```

Подготовим итоговую функцию для расчета и визуализации конверсии

```python
# главная функция расчета и визуализации конверсии

'''
Перед использованием функции небходимо подготовить две таблицы:
1. Таблица профилей (profiles) вида: 
id клиента (user_id) - дата и время регистрации / первого посещения (reg_time) - доп. атрибуты (dimensions)
2. Таблица покупок (payments) вида:
id клиента (user_id) - дата и время покупки (pay_time)
'''

def conversion_final(
    profiles,               # таблица профилей
    payments,               # таблица покупок
    user_id,                # столбец с id клиентов в таблице покупок и профилей
    reg_time,               # столбец с датой регистраций (или первых посещений) в таблице профилей
    pay_time,               # столбец с датой покупок в таблице покупок
    observation_date,       # момент анализа
    horizon_days,           # горизонт анализа 
    dimensions = [],        # группировка по желаемым признакам
    ex_lifetimes = [],      # исключение желаемых лайфтаймов
    comm_axis = False,      # сделать ли общую ось у двух графиков
    title = 'Конверсия',    # заголовок графиков
    window = 1,             # параметр сглаживания графика динамики конверсии
    ignore_horizon = False, # игнорировать ли горизонт
    heatmap_on = False,     # нужно ли также построить хитмеп конверсий по когортам дней регистрации
    heatmap_size = 3,       # задаем размер тепловой карты
    heatmap_annot = True    # включить значения на тепловой карте
):
    
    # получаем таблицы конверсии
    (
        conversion_raw,
        conversion_grouped, 
        conversion_history,
        conversion_heatmap
    ) = get_conversion(
        profiles, 
        payments,
        user_id,
        reg_time,
        pay_time,
        observation_date, 
        horizon_days,
        dimensions,
        ignore_horizon
    )

    # строим графики удержания
    plot_conversion(
        conversion_grouped, 
        conversion_history,
        conversion_heatmap,
        horizon_days,
        dimensions,
        ex_lifetimes,
        heatmap_on,
        comm_axis,
        title,
        window,
        heatmap_size,
        heatmap_annot
    )
```

#### Расчет и визуализация оттока

Подготовим вспомогательную функцию для расчета оттока

```python
# вспомогательная функция для расчёта оттока

def get_churn(
    profiles,           # таблица профилей
    sessions,           # таблица посещений
    user_id,            # столбец с id клиентов в таблице посещений и профилей
    reg_time,           # столбец с датой регистраций (или первых посещений) в таблице профилей
    session_time,       # столбец с датой посещений в таблице посещений
    observation_date,   # момент анализа
    horizon_days,       # горизонт анализа
    dimensions,         # группировка по желаемым признакам
    ignore_horizon,     # игнорировать ли горизонт
):

    # проверяем момент анализа
    if observation_date > sessions[session_time].max():
        print('Внимание! Заданный вами момент анализа превышает максимальную дату посещений. '
              'Значение момента анализа уменьшено до максимально возможного: {}.\n'
              .format(sessions[session_time].max().date()))
        observation_date = sessions[session_time].max().date()
    
    # добавляем дату первых посещений в таблицу профилей
    profiles.loc[:,'dt'] = profiles[reg_time].dt.date
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
        # если нет пользователей, подпадающих под параметры, выводим сообщение об ошибке
        if profiles['dt'].min() > last_suitable_acquisition_date:
            raise Exception('Ошибка! Нет данных, подпадающих под заданные в функции параметры. '
                            'Рекомендуется уменьшить значение горизонта анализа '
                            'и/или увеличить значение момента анализа.')
    # оставляем только нужных пользователей
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # добавляем данные о посещениях в профили
    result_raw = result_raw.merge(
        sessions[[user_id, session_time]], on=user_id, how='left'
    )
    
    # рассчитываем лайфтайм пользователя для каждой посещения
    result_raw['lifetime'] = (
        result_raw[session_time] - result_raw[reg_time]
    ).dt.days.astype('Int64')
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days, k=[0]):
        # строим «треугольную» таблицу посещений
        result = df.pivot_table(
            index=dims, columns='lifetime', values=user_id, aggfunc='nunique'
        )
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result.loc[:, (result.columns < horizon_days)]
        # находим лайфтаймы, за которые нет данных
        set_1 = set(result.columns.astype(int))
        list_all = sorted(list(set_1))
        set_1 = set(filter(lambda x: x >= 0, set_1))
        set_2 = set(range(result.columns.astype(int).max()+1))
        list_nan = sorted(list(set_1 ^ set_2))
        if len(list_nan) > 0:
            # заполняем пропуски в лайфтамах значениями из предыдущих столбцов
            def insert_number(lst, num, df_in):
                index = len(lst)
                for i, n in enumerate(lst):
                    if n > num:
                        index = i
                        break
                    value = n    
                df_in.insert(index, num, df_in[value])
                return df_in
            for lt in list_nan:
                result = insert_number(result.columns.astype(int), lt, result)
            # выводим предупреждение (только если функция вызывается впервые)
            k[0] += 1
            if k[0] == 1:
                list_format = ', '.join(str(x) for x in list_nan)
                print('Внимание! За эти лайфтаймы нет данных: {}'.format(list_format))
                print('Пропуски заполнены значениями из предыдущих лайфтаймов.\n')
        result = 1- result.div(result.shift(1, axis=1))
        return result

    # получаем таблицу оттока
    result_grouped = group_by_dimensions(
        result_raw, dimensions, horizon_days
    )
    
    # для таблицы динамики оттока убираем 'cohort' из dimensions
    if 'cohort' in dimensions:
        dimensions = []

    # получаем таблицу динамики оттока
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )
    
    # получаем таблицу динамики оттока для хитмепа (без группировок по признакам)
    result_heatmap = group_by_dimensions(
        result_raw, ['dt'], horizon_days
    )
        
    # возвращаем таблицы и сырые данные
    return (
        result_raw,       # сырые данные
        result_grouped,   # таблица оттока
        result_in_time,   # таблица динамики оттока
        result_heatmap    # таблица динамики оттока для хитмпепа
    )   
```

Подготовим вспомогательную функцию для визуализации оттока

```python
# вспомогательная функция для визуализации оттока

def plot_churn(
    churn,                 # таблица оттока
    churn_history,         # таблица динамики оттока
    churn_heatmap,         # таблица динамики оттока для хитмепа
    horizon,               # горизонт анализа
    dimensions,            # группировка по желаемым признакам
    ex_lifetimes,          # исключить ненужные лайфтаймы
    heatmap_on,            # нужно ли также строить хитмеп удержания по когортам дней регистрации
    comm_axis,             # сделать ли общую ось у двух графиков
    title,                 # заголовок графиков
    window,                # параметр сглаживания графика динамики оттока
    heatmap_size,          # размер тепловой карты
    heatmap_annot          # включить значения на тепловой карте
):

    # функция для сглаживания фрейма
    def filter_data(df, window):
        for column in df.columns.values:
            df[column] = df[column].rolling(window).mean() 
        return df
    
    # задаём размер сетки для графиков
    plt.figure(figsize=(12, 5))
    
    # задаем стиль графиков
    sns.set_style('darkgrid')

    # если требуется, исключаем заданные лайфтаймы
    if len(ex_lifetimes) != 0:
        churn = churn.drop(columns=ex_lifetimes)
        churn_history = churn_history.drop(columns=ex_lifetimes)
    # в таблице динамики оттока оставляем только нужный лайфтайм
    if churn_history.columns.astype(int).max() == (horizon - 1):
        churn_history = churn_history[[horizon - 1]]
    else:
        # если нужного лайфтайма нет, выводим предупреждение
        print('Внимание! Нет данных за {} лайфтайм. ' 
              'Графики построены до максимального {} лайфтайма.\n'
              .format((horizon - 1),churn_history.columns.astype(int).max()))
    
    # строим график оттока
    ax1 = plt.subplot(1, 2, 1)
    churn.T.plot(
        grid=True, ax=ax1, marker='o'
    )
    plt.xlabel('Лайфтайм')
    plt.title(title)
    plt.gca().yaxis.set_major_formatter('{x:.2%}')
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()

    # строим график динамики оттока
    if comm_axis == False:
        ax2 = plt.subplot(1, 2, 2)
    else:
        ax2 = plt.subplot(1, 2, 2, sharey = ax1)
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in churn_history.index.names
        if name not in ['dt']]
    # фильтруем данные и строим график
    filtered_data = churn_history.pivot_table(
        index='dt', columns=columns, values=churn_history.columns.astype(int).max(), aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax2, marker='o', rot=45)
    plt.gca().yaxis.set_major_formatter('{x:.2%}')
    plt.xlabel('Дата регистрации')
    plt.title('{}: динамика на {}-й день'.format(title,churn_history.columns.astype(int).max()+1))
    if len(dimensions) == 0:
        plt.legend().set_visible(False) 
    else:
        plt.legend()

    # выводим оба графика
    plt.tight_layout()
    plt.show()
    
    # если нужно, также строим хитмеп оттока по когортам дней регистрации
    if heatmap_on != False:
        plt.figure(figsize = (12, churn_heatmap.shape[0]/heatmap_size))
        # если требуется, исключаем заданные лайфтаймы
        if len(ex_lifetimes) != 0:
            churn_heatmap = churn_heatmap.drop(columns=ex_lifetimes)
        sns.heatmap(churn_heatmap,
                annot = heatmap_annot,
                fmt = '.2%', 
                cmap= 'coolwarm',
                linecolor='black',
                linewidths=0.01
               )
        plt.title('{}: тепловая карта'.format(title))
        plt.ylabel('Дата регистрации')
        plt.xlabel('Лайфтайм')
        plt.show()
```

Подготовим главную функцию для расчета и визуализации оттока

```python
# главная функция расчета и визуализации оттока

'''
Перед использованием функции небходимо подготовить две таблицы:
1. Таблица профилей (profiles) вида: 
id клиента (user_id) - дата и время регистрации / первого посещения (reg_time) - доп. атрибуты (dimensions)
2. Таблица посещений (sessions) вида:
id клиента (user_id) - дата и время посещения (session_time)
'''

def churn_final(
    profiles,               # таблица профилей
    sessions,               # таблица посещений
    user_id,                # столбец с id клиентов в таблице посещений и профилей
    reg_time,               # столбец с датой регистраций (или первых посещений) в таблице профилей
    session_time,           # столбец с датой посещений в таблице посещений
    observation_date,       # момент анализа
    horizon_days,           # горизонт анализа 
    dimensions = [],        # группировка по желаемым признакам
    ex_lifetimes = [],      # исключить ненужные лайфтаймы
    comm_axis = False,      # сделать ли общую ось у двух графиков
    title = 'Удержание',    # имя переменной в заголовках графиков
    window = 1,             # параметр сглаживания графика динамики оттока
    ignore_horizon = False, # игнорировать ли горизонт
    heatmap_on = False,     # нужно ли также построить хитмеп оттока по когортам дней регистрации
    heatmap_size = 3,       # задаем размер тепловой карты
    heatmap_annot = True    # включить ли подписи на тепловой карте
):
    
    # получаем таблицы оттока
    (
        churn_raw,
        churn_grouped,
        churn_history,
        churn_heatmap
    ) = get_churn(
        profiles,
        sessions,
        user_id,
        reg_time,
        session_time,
        observation_date,
        horizon_days,
        dimensions,
        ignore_horizon
        
    )
    
    # строим графики оттока
    plot_churn(
        churn_grouped,
        churn_history,
        churn_heatmap,
        horizon_days,
        dimensions,
        ex_lifetimes,
        heatmap_on,
        comm_axis,
        title,
        window,
        heatmap_size,
        heatmap_annot
    )
```

Посчитаем, как платежи распределяются по уровням.

```python
# готовим таблицу платежей
p = payments.rename(columns={'pay_time':'time'})

# готовим таблицу уровней
l = levels.rename(columns={'level_time':'time'})

# соединяем вертикально таблицы платежей и уровней
p_l = pd.concat([p, l], ignore_index=True).sort_values(by=['user_id','time']).reset_index(drop=True)

# для каждого платежа находим уровень, на котором он прошел
p_l['level'] = p_l.groupby('user_id')['level'].fillna(method='ffill')
       
# заполняем пропуски в уровнях 1-м уровнем для платежей, прошедших до 2 уровня
p_l['level'] = p_l['level'].fillna(1.0).astype(int)

# находим суммарную выручку на каждом уровне
rev_lev = p_l.groupby('level')['amt'].sum().sort_index()

# строим график
rev_lev.plot(kind='bar')
plt.title('Суммарная выручка по уровням')
plt.show()
```

По какой-то причине именно на 8 уровне у учеников появляется сильное желание платить. Но похоже, сложность прохождения дальнейших уровней может влиять на удержание учеников, что уменьшает потенциальную выручку с этих уровней. Рекомендуется уменьшить сложность прохождения уровней после 8 уровня, чтобы уменьшить отток учеников после 8 уровня, тем самым повысив выручку.
- Больше всего пользователей отваливается с 1-го на 2-й день использования приложения (с 0-го на 1-й лайфтайм). Рекомендуется внедрить способы удержания пользователей на 2-й день (напоминание в приложении, предоставление скидки и т.д.)

## Ответы на вопросы

**1. Дайте определение ARPPU. В результате изменений в продукте ARPPU снизился. Это хорошо или плохо? Отвечайте развернуто.**

ARPPU - это средний доход с платящего пользователя. Находится путем деления дохода с платящих пользователей на число платящих пользователей.

Если ARPPU снизился - это плохо: те пользователи, которые вообще готовы платить, стали меньше платить, а значит выручка с когорты снизилась. Однако, если при этом конверсия в платящих пользователей увеличилась, то в целом выручка могла остаться такой же или даже увеличиться. Грубо говоря, если ARPPU снизился, но конверсия увеличилась, то качество платящих пользователей уменьшилось, но количественно их стало больше. Поэтому, чтобы понять, какое влияние на выручку оказывает снижение ARPPU, само по себе изменение ARPPU не стоит рассматривать отдельно, стоит также смотреть на конверсию.

**2. Предположим, что в результате обновления дизайна продукта вы наблюдаете
увеличение среднего времени нахождения пользователей в приложении, но
уровень конверсии в покупку снизился. Как вы будете анализировать данное
изменение и какие рекомендации вы предложите команде продукта?**

Если изменение дизайна уже произошло, то буду анализировать данные изменения путем построение графиков среднего времени в приложении на одного пользователя по дням с накоплением и построения графика конверсии в покупателей по дням с накоплением. Если когорты новых привлеченных пользователей покажут на заданный лайфтайм (например, 14-й лайфтайм), явное увеличение времени в приложении и уменьшения конверсии по сравнению тем же лайфтаймом у когорт, привлеченных до изменения дизайна, то сообщу об этом команде продукта.

Если после изменения дизайна среднее время увеличилось, а конверсия уменьшилась, то можно рекомендовать команде продукта исследовать то, не мешает ли новый дизайн совершать покупки (сложности оформления заказа, скрытие нужных элементов интерфейса т.д.) - это можно сделать, например, через анализ воронки в покупки (посмотреть, на каком шаге пользователи стали сильнее отваливаться после изменения дизайна). Если изменения конверсии не удастся исправить, то для компенсации уменьшения выручки от уменьшения конверсии можно рекомендовать: увеличить удержание пользователей, увеличить ARPPU.

Если изменение дизайна еще не произошло, но предполагается в будущем, предложу сначала провести A/B тест. Показать контрольной группе - старый дизайн, экспериментальной - новый. Взять время в приложении на 14 день для каждого пользователя и провести стат. тест (например, т-тест) для определения того, статистистически ли значима разница во времени в двух выборках и величины этой разницы. Также проведу z-тест на равенство долей для определения того, есть ли разница в конверсиях в покупателей в двух выборках на 14 день. По итогам сделаю выводы об успешности задумки об изменения дизайна.

**3. Выберете 3-4 метрики, которые следует отслеживать в первую очередь? Какие еще метрики следует отслеживать? Какие данные стоит собирать дополнительно к уже собираемым, чтобы рассчитать предлагаемые тобою метрики?**

В нашем проекте стоит в первую очередь отслеживать LTV, конверсию, удержание. Однако сам по себе LTV в отрыве от САС (стоимости привлечения) не дает информации о том, какую прибыль мы получаем от клиентов. Поэтому считаю, что стоит также отслеживать ROI (возврат на инвестиции) = LTV / САС, который будет более показателен для анализа прибыли за вычетом стоимости привлечения: покажет насколько окупаются привлеченные пользователи. Для его подсчета требуются дополнительные данные по стоимости привлечения пользователей.
